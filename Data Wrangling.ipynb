```python
""" Printing unique values of all object-type columns """

print('Gender:              ', churnData.gender.unique())
print('Partner:             ', churnData.Partner.unique())
print('Dependents:          ', churnData.Dependents.unique())
print('PhoneService:        ', churnData.PhoneService.unique())
print('MultipleLines:       ', churnData.MultipleLines.unique())
print('InternetService:     ', churnData.InternetService.unique())
print('OnlineSecurity:      ', churnData.OnlineSecurity.unique())
print('OnlineBackup:        ', churnData.OnlineBackup.unique())
print('DeviceProtection:    ', churnData.DeviceProtection.unique())
print('TechSupport:         ', churnData.TechSupport.unique())
print('StreamingTV:         ', churnData.StreamingTV.unique())
print('StreamingMovies:     ', churnData.StreamingMovies.unique())
print('Contract:            ', churnData.Contract.unique())
print('PaperlessBilling:    ', churnData.PaperlessBilling.unique())
print('PaymentMethod:       ', churnData.PaymentMethod.unique())
print('Churn:               ', churnData.Churn.unique())
```

```python
""" Binary Encoding gender column """

churnData['gender'] = churnData.gender.astype('category').cat.codes # Female = 0 and Male = 1
```

```python
""" Binary Encoding Partner column """

categoriesPartner = ['No','Yes']
Dtype = CategoricalDtype(categories=categoriesPartner,ordered=True)
churnData.Partner = churnData.Partner.astype(Dtype).cat.codes # No = 0 and Yes = 1
```

```python
""" Binary Encoding Dependents column """

categoriesDependents = ['No','Yes']
Dtype = CategoricalDtype(categories=categoriesDependents,ordered=True)
churnData.Dependents = churnData.Dependents.astype(Dtype).cat.codes # No = 0 and Yes = 1
```

```python
""" Binary Encoding PhoneService Column """

categoriesPhoneService = ['No','Yes']
Dtype = CategoricalDtype(categories=categoriesPhoneService,ordered=True)
churnData.PhoneService = churnData.PhoneService.astype(Dtype).cat.codes                 # no = 0 and yes = 1
```

```python
""" Label Encoding MultipleLines column """

categoriesMultipleLines = ['No phone service', 'No','Yes']
Dtype = CategoricalDtype(categories=categoriesMultipleLines,ordered=True)
churnData.MultipleLines = churnData.MultipleLines.astype(Dtype).cat.codes                 # No phone service = 1, No=1, Yes = 2
```

```python
""" Label Encoding InternetService Column """

categoriesInternetService = ['No','DSL','Fiber optic']
Dtype = CategoricalDtype(categories=categoriesInternetService,ordered=True)
churnData.InternetService = churnData.InternetService.astype(Dtype).cat.codes           # No=0,DSL=1,Fiber optic=2
```

```python
""" Label Encoding OnlineSecurity Column """

categoriesOnlineSecurity = ['No internet service', 'No', 'Yes']
Dtype = CategoricalDtype(categories=categoriesOnlineSecurity,ordered=True)
churnData.OnlineSecurity = churnData.OnlineSecurity.astype(Dtype).cat.codes             # No internet service=0, No=1, Yes=2
```

```python
""" Label Encoding OnlineBackup Column """

categoriesOnlineBackup = ['No internet service', 'No', 'Yes']
Dtype = CategoricalDtype(categories=categoriesOnlineBackup,ordered=True)
churnData.OnlineBackup = churnData.OnlineBackup.astype(Dtype).cat.codes                 # No internet service=0, No=1, Yes=2
```

```python
""" Label Encoding DeviceProtection Column """

categoriesDeviceProtection = ['No internet service', 'No', 'Yes']
Dtype = CategoricalDtype(categories=categoriesDeviceProtection,ordered=True)
churnData.DeviceProtection = churnData.DeviceProtection.astype(Dtype).cat.codes         # No internet service=0, No=1, Yes=2
```

```python
""" Label Encoding TechSupport Column """

categoriesTechSupport = ['No internet service', 'No', 'Yes']
Dtype = CategoricalDtype(categories=categoriesTechSupport,ordered=True)
churnData.TechSupport = churnData.TechSupport.astype(Dtype).cat.codes                   # No internet service=0, No=1, Yes=2
```

```python
""" Label Encoding StreamingTV Column """

categoriesStreamingTV = ['No internet service', 'No', 'Yes']
Dtype = CategoricalDtype(categories=categoriesStreamingTV,ordered=True)
churnData.StreamingTV = churnData.StreamingTV.astype(Dtype).cat.codes                   # No internet service=0, No=1, Yes=2
```

```python
""" Label Encoding StreamingMovies Column """

categoriesStreamingMovies = ['No internet service', 'No', 'Yes']
Dtype = CategoricalDtype(categories=categoriesStreamingMovies,ordered=True)
churnData.StreamingMovies = churnData.StreamingMovies.astype(Dtype).cat.codes            # No internet service=0, No=1, Yes=2
```

```python
""" One-hot Encoding PaymentMethod & Contract Columns """

columnsOneHot = ['PaymentMethod', 'Contract']
churnData=pd.get_dummies(churnData,columns=columnsOneHot)
```

```python
""" Binary Encoding PaperlessBilling Column """

categoriesPaperlessBilling = ['No','Yes']
Dtype = CategoricalDtype(categories=categoriesPaperlessBilling,ordered=True)
churnData.PaperlessBilling = churnData.PaperlessBilling.astype(Dtype).cat.codes           # No=0,Yes=1
```

```python
""" Binary Encoding Churn Column """
categoriesChurn = ['No','Yes']
Dtype = CategoricalDtype(categories=categoriesChurn,ordered=True)
churnData.Churn = churnData.Churn.astype(Dtype).cat.codes                                  # No=0,Yes=1
```
